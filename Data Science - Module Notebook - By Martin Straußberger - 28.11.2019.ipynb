Clinically, EEG is most commonly referred to diagnose sleep disorders, epilepsy, coma, brain death and more. Today, thanks to companies like Muse and Emotiv and their Developer Kits, the benefits of EEG are now publicly available. For example, to help people improving their focus, meditation, sleep and even to control and play video games or VR sceneries.

Many different plots were used to display and compare both Datasets to eventually conclude.

## 1. Import Dependencies

```python
import numpy as np
from numpy import array
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Loading Data of Subject 1:

```python
subject1 = pd.read_csv("Subject1_eegData.csv")
subject1
```

## 3. Data Cleaning
**3.1 Dropping irrelevant columns.**

```python
subject1 = subject1.drop(['ThetaAF8', 'ThetaAF7', 'ThetaTP9', 'ThetaTP10','DeltaTP9', 'DeltaTP10', 'DeltaAF8', 'Beta_AF8', 'Gamma_AF8', 'Alpha_AF8'], axis=1)
subject1
```

**3.2 Drop NaN's in each column.**

```python
subject1.dropna(how='any', inplace=True)
subject1.tail()
```

**3.3 Apply function to replace dots.**

*The dots cause the values to be read as a string. Converting to numeric.*

```python
f = lambda X : X.replace('.', '')

columns = subject1.columns
for cols in columns : 
    subject1[cols] = subject1[cols].apply(f)

subject1
```

## 4. Data Converting:
**4.1 Describe dtypes and converting to floats.**

```python
subject1.dtypes
```

```python
cols=[i for i in subject1.columns if i not in ["TimeStamp"]]
for col in cols:
    subject1[col]=pd.to_numeric(subject1[col], errors = 'coerce')
    
subject1.dtypes
```

**4.2 Rename relevant columns and converting from int64 to float64.**

```python
sj1_BetaTP9 = subject1['Beta_TP9'].astype(np.float64)
sj1_BetaAF7 = subject1['Beta_AF7'].astype(np.float64)
sj1_GammaTP9 = subject1['Gamma_TP9'].astype(np.float64)
sj1_GammaAF7  = subject1['Gamma_AF7'].astype(np.float64)
```

**4.3 Drop NaN caused by converting to floats with Errorhandling of errors = 'coerce'.**

```python
subject1.dropna(how='any', inplace=True)
subject1.describe
```

**4.4. Checking values from relevant Columnns of Dataset from Subject 1:**

```python
sj1_BetaTP9.dtypes, sj1_BetaAF7.dtypes, sj1_GammaTP9.dtypes, sj1_GammaAF7.dtypes
```

## 5. Loading Data from Subject 2:

```python
subject2 = pd.read_csv("pre_processed_eegSubject2")
```

**5.1 Naming the relevant Columns**

```python
sj2_BetaTP9 = subject2['Beta_TP9']
sj2_BetaAF7 = subject2['Beta_TP9'] # prefrontal cortex
sj2_GammaAF7 = subject2['Gamma_AF7'] # prefrontal cortex
sj2_GammaTP9 = subject2['Gamma_TP9'] # left hemisphere behind the ear
```

**5.2 Re-check Values and dtypes of Columns from Subject 2:**

```python
sj2_BetaTP9, sj2_BetaAF7, sj2_GammaAF7, sj2_GammaTP9
```

**5.3 Saving columns as float64**

```python
sj2_BetaTP9 = sj2_BetaTP9.astype(np.float64)
sj2_BetaAF7 = sj2_BetaTP9.astype(np.float64)
sj2_GammaAF7 = sj2_GammaAF7.astype(np.float64)
sj2_GammaTP9 = sj2_GammaTP9.astype(np.float64)
```

**5.4 Checking the shape of columns**

```python
sj2_BetaTP9.shape, sj2_BetaAF7.shape, sj2_GammaAF7.shape, sj2_GammaTP9.shape
```

## 6. Dependency: Import Seaborn to visualize both Datasets

```python
import seaborn as sns
sns.set(font_scale=1.1)
```

### 6.1 Visualize β-Waves from Subject 1:
In the following parts of this Notebook from **6.1 to 6.4** I separated both subjects. The main purpose of separating both is to first understand each states. 

Further, since **β-, and  γ-, Waves** are **relevant for focusing, learning as well as processing information** I will first visualize:
1. β-Waves
2. γ-, Waves
3. Finally, both wave types overlapped, which serves as a bigger picture.

```python
#BetaAF7 of Subject 1
sf = 1.4
time = np.arange(sj1_BetaAF7.size) / sf

fig, ax = plt.subplots(1, figsize=(12,4))
plt.plot(time, sj1_BetaAF7, lw=2.0, color='salmon')
plt.xlabel('Time (seconds)')
plt.ylabel('µV²/Hz (dB)')
plt.xlim([time.min(), time.max()])
plt.title('Beta Channel: AF7')
sns.despine()

chartinfo = ('Fig.1: β-Waves Sj. 1')
plt.annotate(chartinfo, xy=(0.0, -0.2),size=17, xycoords='axes fraction')



#GammaAF7 of Subject 1
sf = 1.4
time = np.arange(sj1_GammaAF7.size) / sf

fig, ax = plt.subplots(1, figsize=(12,4))
plt.plot(time, sj1_GammaAF7, lw=2.0, color='b')
plt.xlabel('Time (seconds)')
plt.ylabel('µV²/Hz (dB)')
plt.xlim([time.min(), time.max()])
plt.title('Gamma Channel: AF7')
sns.despine()


chartinfo = ('Fig.2: γ-Waves Sj. 1')
plt.annotate(chartinfo, xy=(0.0, -0.2),size=17, xycoords='axes fraction')
```

**Fig.1 and Fig.2** are displaying the differences in between β-, and  γ-, Waves of Subject 1.

## 6.2 Plot overlapped  β-, and  γ-, Waves of Subject 1:

```python
#BetaAF7
sf = 100.
time = np.arange(sj1_BetaAF7.size) / sf
#GammaAF7
sf = 100.
time = np.arange(sj1_GammaAF7.size) / sf


fig, ax = plt.subplots(1, figsize=(12,4))
plt.plot(time, sj1_BetaAF7, lw=2.0, color='salmon')
plt.plot(time, sj1_GammaAF7, lw=2.0, color='b')
plt.xlabel('Time (seconds)')
plt.ylabel('µV²/Hz (dB)')
plt.xlim([time.min(), time.max()])
plt.title('Beta and Gamma combined Channel: AF7')
sns.despine()


chartinfo = ('Fig.3: Sj. 1 - Overlapping  β-Waves and  γ-Waves ')
plt.annotate(chartinfo, xy=(0.0, -0.3),size=17, xycoords='axes fraction')
```

# 7. Comparing β-Waves of both Subjects :

```python
plt.figure(figsize=(18,4));
# Graph Size
plt.subplot(121)
# Plotting Betawaves from Subject 1
plt.plot(sj1_BetaAF7, 'bo', sj1_BetaAF7, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('Beta Channel: AF7');

chartinfo = ('Fig.4: Dashed Density of β-Waves Sj. 1')
plt.annotate(chartinfo, xy=(0.0, -0.2),size=17, xycoords='axes fraction')


plt.figure(figsize=(18,4));
# Graph Size
plt.subplot(121)
# Plotting Betawaves from Subject 2
plt.plot(sj2_BetaAF7, 'bo', sj2_BetaAF7, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('Beta Channel: AF7');

chartinfo = ('Fig.4.1: Dashed Density of β-Waves Sj. 2')
plt.annotate(chartinfo, xy=(0.03, -0.2),size=17, xycoords='axes fraction')
```

A **valuable method** to **grasp a bigger picture** of each individual's mind-state is to plot the **waves in dashes** as in **Fig.4 and Fig.4.1**, by making the surface in between the dots invisible by assigning **'mo'** to the plot. It **uncovers a good view** to get an idea of the waves if they are scattered, spread or show common patterns.

Within the scope of measurement accuracy the **β-Waves** and it's dense of the **dots** were mainly **close to zero, supports the fact** that both were increased consciousness and **processing information**.

## 7.1 Plot overlapped β-Waves of both Subjects for direct comparison: 

```python
plt.figure(figsize=(18,4));
# Graph Size
plt.subplot(121)
# Plotting Overlapping Graphs for Visualization
plt.plot(sj2_BetaAF7, 'black', sj2_BetaAF7, 'c', alpha=.25);
plt.plot(sj1_BetaAF7, 'black', sj1_BetaAF7, 'crimson', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('Concentration Frame of β-Waves: Subject 1 + Subject 2');

chartinfo = ('Fig.4.2: Comparing β-Waves of Sj. 2 and Sj. 1')
plt.annotate(chartinfo, xy=(0.03, -0.2),size=17, xycoords='axes fraction')
```

After identifying the most frequent levels of the brainwaves as highlighted with **Fig.4 and Fig.4.1**, **Fig.4.2** reveals that Subject 2 was slightly more active within the β-Waves, whereas Subject 2 was also frequent but less noisy (lower peaks on average). Those **graphs** are **emphasizing** the fact that **both** subjects **were clearly processing information**.

# 9. Density Spotting of Brainwaves from both Subjects:
**10.1 Adjusting parameters for plots.**

```python
sj1_BetaAF7 = np.random.normal(sj1_BetaAF7)
sj1_GammaAF7 = np.random.normal(sj1_GammaAF7)

sj2_BetaAF7 = np.random.normal(sj2_BetaAF7)
sj2_GammaAF7 = np.random.normal(sj2_GammaAF7)
```

# 10. Density spotting β-Waves: Jointplots for Subject 1 and Subject 2

```python
#BetaAF7 Subject 1:
with sns.axes_style('whitegrid'):
    
    sns.jointplot(sj1_BetaAF7, sj1_BetaAF7, kind="kde", color="lightblue", ratio= 2,
                  marginal_kws={}).set_axis_labels('scale','µV²/Hz (dB)');
    
    plt.text(0.0,0.0, "Density: β-Waves Subject 1", fontsize= 25, color='Grey', horizontalalignment='left',
             verticalalignment='baseline', rotation=0);
    
    chartinfo = ('Fig.6:')
    plt.annotate(chartinfo, xy=(-2.6, -0.2),size=17, xycoords='axes fraction')

#BetaAF7 Subject2:
with sns.axes_style('whitegrid'):
    
    sns.jointplot(sj2_BetaAF7, sj2_BetaAF7, kind="kde", color="lightblue", ratio= 2,
                  marginal_kws={}).set_axis_labels('scale','µV²/Hz (dB)');
    
    plt.text(0.0,0.0, "Density: β-Waves Subject 2", fontsize= 25, color='Grey', horizontalalignment='left',
             verticalalignment='center', rotation=0);
    
    chartinfo = ('Fig.6.1')
    plt.annotate(chartinfo, xy=(-2.6, -0.2),size=17, xycoords='axes fraction')
```

The direct comparison of **β-Waves** between both Subjects displays clearly that there are differences of the **individual β-Wave levels.** Seeing both Figures *(Fig.6 and Fig.6.1)* allows us to already make first statements:

**1.** The **β-Waves of Subject 1** show a **high activity** close to zero (baseline), ranging up to **+/- 0.5**

**2.** The **β-Waves of Subject 2** show **also** a **high activity** close to the zero (baseline), with less high but denser peaks approximately up to **+/-0.3**, **but 2 density spots**. One closer to zero and seconds amplitudes are in between **0.7 - 1.0**

# 10.1 Density spotting γ-Waves: Jointplots for Subject 1 and Subject 2

# 11. Peak Comparison of γ-, and β- Waves: Both Subjects

```python
import seaborn as sborn
sborn.set(font_scale=1.1)

sns.kdeplot(sj1_BetaAF7, color='lightblue',   label = 'sj1_BetaAF7');
sns.kdeplot(sj1_GammaAF7, color='mediumblue', label = 'sj1_GammaAF7');
sns.kdeplot(sj2_BetaAF7, color='crimson', label = 'sj2_BetaAF7');
sns.kdeplot(sj2_GammaAF7, color='orange', label = 'sj2_GammaAF7');


chartinfo = ('Fig.8: Direct Comparison of Subject 1 and Subject 2')
plt.annotate(chartinfo, xy=(0, -0.3),size=17, xycoords='axes fraction')
```

Lastly using the method of plotting with **sns.kdeplots** provides a visualization of their wave peaks. 

It helps to identify the thresholds differences of both subjects which are displayed in **Fig.8**.

**In science highly active γ-Waves are stated as evidence of a very focused brain with an increased capability of deep learning and deep meditation.** Whereas β- Waves are associated with a conscious mind, which means thinking and decision making. β-Waves, therefore play also an essential role in how to interpret someone's level of concentration.

With **Fig. 8**, it is arguable that both subjects are highly focused, but Subject 1 displays a slightly more γ-Waves activity which is under consideration of the conditions of this experiment a hint to make the assumption that Subject 1 was more focused or deeper in processeing the read information of the book.
